```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Loading the dataset
credit_dataset = pd.read_csv('/content/creditcard.csv')
credit_dataset.shape
```

```python
credit_dataset.head()
```

```python
credit_dataset.describe()
```

```python
print(credit_dataset['Class'].value_counts())
```

```python
print(credit_dataset.info())
```

```python
credit_dataset.groupby(['Class']).mean()
```

This Dataset is highly unbalanced

0 --> Nornal transaction

1 --> Fraud transaction

```python
# separating the data for analysis
legit = credit_dataset[credit_dataset.Class==0]
fraud = credit_dataset[credit_dataset.Class==1]
```

```python
# statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
sample_legit = legit.sample(n=403)  # placing random 403 sample from legit to sample_legit
```

```python
new_credit_data = pd.concat([sample_legit,fraud],axis=0)
new_credit_data.head()
```

```python
new_credit_data.shape
```

```python
new_credit_data.groupby('Class').mean()
```

Splitting the dataset

```python
X = new_credit_data.drop(columns='Class',axis=1)
Y = new_credit_data['Class']
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

model training

```python
model = LogisticRegression()
model.fit(X_train,Y_train)
```

Model Evaluvation

```python
training_data_predict = model.predict(X_train)
accuracy = accuracy_score(training_data_predict,Y_train)
print("Accuracy on training data",accuracy)
```

```python
test_data_predict = model.predict(X_test)
accuracy = accuracy_score(test_data_predict,Y_test)
print("Accuracy on test data",accuracy)
```
